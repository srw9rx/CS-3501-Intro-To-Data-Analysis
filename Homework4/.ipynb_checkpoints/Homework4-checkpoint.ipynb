(b) Perform the steps outlined above to find the optimal rotation that aligns hand 2 into
hand 1. What is the angle of rotation between these two hands?

```python
#create the matrix

#hand1t * hand2
hand1np = hand1.to_numpy()
#onetranspose = hand1np.transpose()
hand2np = hand2.to_numpy()
multiplied = np.multiply(hand1np, hand2np)
```

```python
#U*S*VT = A1T*A2
#s diagonal matrix 
u,s,vt =  svd(multiplied)
```

```python
#angle 
#multiplied divided by s
#r = np.dot(u, vt)
r1 = u[0][0] * vt[0][0]
r2 = u[0][1] * vt[0][1]
r3 = u[1][0] * vt[1][0]
r4 = u[1][1] * vt[1][1]

angle = np.arccos(r1)
angledeg = np.rad2deg(angle)
print(angledeg)
```

(c) Rotate hand 2 using the optimal rotation you found. Plot the two hands again (this
time with hand 2 rotated). Do they align with each other?

```python
import math
xar = []
yar = []
for i in range(0, 72, 1):
    x,y = hand2np[i][0], hand2np[i][1]
    x = x*math.cos(angle) + y*math.sin(angle)
    y = y*math.cos(angle) - x*math.sin(angle)
    xar.append(x)
    yar.append(y)
    #np.append(hand2rot, y, axis=1)
#hand2rotl.pop(0)
#print(hand2rotl)
xseries = pd.Series(xar)
yseries = pd.Series(yar)
print(xseries)
```

2. Now we are going to use Principal Component Analysis to find the most relevant dimensions
of variance in a set of hand shapes. Load the matrix all-hands.dat. Each row of this matrix

Do the following:

(a) Compute the mean hand (this should be a 72 × 2 = 144 vector, consisting of the means
of each column in your matrix). Plot this mean as a hand shape.

```python
#create the data frame
colnames = []
for i in range (1, 73, 1):
    colnames.append("x"+str(i))
    colnames.append("y"+str(i))
allhands= pd.read_table('/Users/sophiawalton/Desktop/CS3501/Homework4/all-hands.dat', sep=",", names = colnames ,header=None)
```

```python
#find the means:
xmeans = []
ymeans = []

#loop through and find each mean
for i in range(1, 72, 1):
    xmean = allhands["x"+str(i)].mean()
    ymean = allhands["y"+str(i)].mean()
    xmeans.append(xmean)
    ymeans.append(ymean)
#create data frame of x and y means

xseries = pd.Series(xmeans)
yseries = pd.Series(ymeans)
```

```python
#plot the mean hand data
x = xseries.to_numpy()
y = yseries.to_numpy()

#plot datapoints 
fig, mean = plt.subplots(figsize=(10, 6))

mean.plot(x, y, color ='purple', marker = 'o')

plt.title("Mean Hand")

plt.show()
```

(b) Compute the covariance matrix Σ for this data. Use the formulas we covered in class,
not a covariance function! What is the total variance of the data?

```python
#compute the means for each - do by looping through
n = 18 #18 rows of data 
centereddf = pd.DataFrame()
for i in range(1,73, 1):
    newxseries = pd.Series()
    newyseries = pd.Series()
    xseries = allhands["x"+str(i)].squeeze()
    yseries = allhands["y"+str(i)].squeeze()
    xmean = allhands["x"+str(i)].mean()
    ymean = allhands["y"+str(i)].mean()
    for uncentered in xseries: 
        newitem = uncentered-xmean
        newxseries.append((1/n)*newitem)
    for uncentered in yseries:
        newitem = uncentered-ymean
        newyseries.append((1/n)*newitem)
    centereddf['x'+str(i)]=newxseries
    centereddf['y'+str(i)]=newyseries
        
    
```
